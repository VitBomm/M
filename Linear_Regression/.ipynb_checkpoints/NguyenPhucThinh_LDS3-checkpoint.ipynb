# Câu 3 Sử dụng Gradient Descent tìm w.
- Thêm 1 cột Prediction2 

```python
y_GS = y.copy().to_numpy().reshape(y.shape[0],1)
```

```python
X_GS = df_train[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY']].to_numpy()
```

# Khởi tạo theta theta = [1, 1, 1, 1]

```python
theta = np.ones(4).reshape(4,1)
```

```python
theta
```

```python
alpha = 0.001
num_iter = 1000
```

# Khởi tạo hàm tính khoảng cách giữa các điểm và theta được truyền vào

```python
# X.shape(854 x 4) and y.shape(4 x 1) => h(854 x 1) - y.shape(854x1) = loss (hi - yi) shape 854 x1
def Compute_Loss(X_GS, y_GS, theta):
    m = X_GS.shape[0]
    h = X_GS.dot(theta)
    loss = h - y
    return loss.to_numpy()
```

# Define Gradient descent

```python
def Gradient_descent(alpha, X_GS, y_GS, num_iter):
    m = X_GS.shape[0]
    theta = np.ones(X_GS.shape[1]).reshape(X_GS.shape[1],1)
    for i in range(num_iter):
        loss = Compute_Loss(X_GS, y_GS, theta)
        J = np.sum(loss ** 2) / (2 * m)
        # X_GS.T.shape (4, 854) loss.shape 854x1 Lấy từng dòng trong X_GS.T nhân cho cột của loss tương đương với việc 
        # đi cộng từ các của loss = > (X_GS.T.dot(loss) / m) trả về 4x1. theta cũng là 4x1 tao đồng loạt update theta
        theta = theta - alpha * (X_GS.T.dot(loss) / m)
    return theta
    
```

```python
theta = Gradient_descent(alpha, X_GS, y_GS, num_iter)
```

```python
theta
```

```python
X_GS_test = df_test[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY']].to_numpy()
```

```python
X_GS_test = np.c_[np.ones(X_GS_test.shape[0]), X_GS_test]
```

```python
prediction2 = X_GS_test.dot(theta)
```
